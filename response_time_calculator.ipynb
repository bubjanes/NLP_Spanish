# Response Time Calculator

```python
import pandas as pd
import re
from dateutil.parser import parse
```

```python
df = pd.read_csv('new_mattermost.csv', sep=',', encoding='utf-8', parse_dates=True, dayfirst=True)
df.sort_values("CreateAt", inplace=True)
new_df = df.loc[:,["CreateAt","Mensaje"]]
new_df.head()
```

```python
rawtimes = new_df.CreateAt.dropna()
print(rawtimes[:25])
```

```python
resp_time = []
# a starting point
time_y = parse('2019-04-15 17:52:09+00:00')
# this loop creates computes the response time
for i in rawtimes:
    time_x = parse(i)  
    time_z = time_x-time_y
    resp_time.append(time_z)
    time_y = time_x
```

```python
# add out new column for response time
new_df["ResponseTime"] = resp_time
```

```python
order = ['CreateAt','Mensaje','ResponseTime']
response_time = new_df[['CreateAt','ResponseTime','Mensaje']]
response_time[:25]
```
